```python
import requests as rq
from bs4 import BeautifulSoup as bs
origin=rq.get('https://www.ptt.cc/bbs/Food/M.1452824583.A.E1D.html',verify=False)
food=bs(origin.text)
for entry in food.select('#main-content'):
    for i in [3,0,2]:
        print(entry.select('.article-meta-value')[i].text)
```
